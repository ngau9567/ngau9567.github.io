## Task 1.B
Crime incidents are reported down to the hour/minute, while stops are only reported down to the day.

Please create a new column based on the `REPORT_DAT` column of the crime incidents data that rounds the exact timestamp for the crime down to the day - for instance, "2017-06-03T12:54:27.000Z" would become "2017-06-03". Call this column `report_daily`.

````python
# Creates a function to extract the year-month-day format by
# locating the index from the beginning to position of 9, which is [:10]
def func_one (input_value):
    new_input = input_value[:10]
    return new_input

# Creates a new column 'report_daily' to apply the function and store the 
# year-month-day format from column 'REPORT_DAT'
crimes_df['report_daily'] = crimes_df['REPORT_DAT'].apply(func_one)
crimes_df.head()
````

## Task 1.C 
Create a new data set `crimes_by_day` that gives the total count of crime incidents per day. For the purposes of aggregating, you can consider each unique `OBJECTID` as reflecting a different crime incident.

Similarly, create a new data set `stops_by_day` that gives the total count of traffic stops per day. For the purposes of aggregating, you can consider each unique `stop_id` as reflecting a different stop.

````python
# creates a pandas dataframe 'crimes_by_day' through grouping by 
# the 'report_daily' column from dataframe 'crimes_df'.
# Then uses the size() function to count the numbers of crime incidents
# whcih is the 'OBJECTID' in the dataframe
crimes_by_day = pd.DataFrame(crimes_df.groupby(['report_daily'])['OBJECTID'].size())

# resets the index so that data frame looks organized, and displays the first five rows
crimes_by_day = crimes_by_day.reset_index()
crimes_by_day.rename(columns = {'OBJECTID':'OBJECTID_Count'}, inplace = True)
crimes_by_day.head()
````

````python
# Similar approaches for 'stops_by_day', creates a pandas dataframe then group by 
# the 'stop_date' column from dataframe 'stops_by_day'.
# Then uses the size() function to count the numbers of traffic stops by day
# whcih is the 'stop_id' in the dataframe
stops_by_day = pd.DataFrame(stops_df.groupby(['stop_date'])['stop_id'].size())

# resets the index so that data frame looks organized, and displays the first five rows
stops_by_day = stops_by_day.reset_index()
stops_by_day.head()
````

## Task 1.D
For either the traffic stops by day or crime incidents by day, check if there are days missing in the period measured. If so, fill them with the mean across all other days for that data.


````python
def insert_row(idx, df, df_insert):
    index=np.arange(idx+1)
    dfA = pd.DataFrame(df.iloc[:idx, ], index=index)
    dfB = df.iloc[idx:, ]
    
    dfA.iloc[-1] = df_insert
    df = dfA.append(dfB).reset_index(drop = True)
    return df
````

````python
# import useful packages datetime
from datetime import datetime
from datetime import date, timedelta

# In this case, I will perform the testing on the traffic stops dataframe
# to find any missing date and fill them with mean value of number of traffic stops across all other days. 
# first, to create a function to convert the string of date format into timestemp format.
# it will helpful to calculate the forward days in the later stage by doing it. 
def func_two (input_value):
    my_date = datetime.strptime(input_value, "%Y-%m-%d")
    return my_date

# creates another function to manipulate the dataframe when the missing date is detected.
# the function takes the index of the missing date, the original dateframe, and the missing date 
# that needs to be appended. The function first split the orginal dataframe into 2 half, the first 
# half has all rows from begining to the index where the missing date found, then the function appends
# the missing date dataframe, and finally appends the second half of the original dataframe which 
# has all the rows from the index of missing date to the end row of the original datafram.
def insert_row(idx, df, df_insert):
    dfA = df.iloc[:idx, ]
    dfB = df.iloc[idx:, ]
    df = dfA.append(df_insert).append(dfB).reset_index(drop = True)
    return df

# creates a new column to store by converting the date column into timestemp format
# by applying the 'func_two' function just created.
stops_by_day['new_date'] = stops_by_day['stop_date'].apply(func_two)

# converts the newly created column 'new_date' into a list
days = stops_by_day['new_date'].to_list()

# calculates the mean value of number of traffic stops across all other days 
mean_value = stops_by_day['stop_id'].mean()

# creates a for loop to go through the list, with the range of number of days by using the last
# day minus the first day using 'days'. Then uses the timedelta function to get one forward date each
# time by adding to the first date. For each forward date, checks it if it exists in the original 
# date list. If yes, keep on the loop. If no, creates the forward date with dataframe along with the 
# the mean value as imput value. Applies to 'insert_row' function to get the new dataframe of 'stops_by_day'.
for i in range((days[-1] - days[0]).days):
    a = days[0] + timedelta(i)
    if a not in days:
        aa =  pd.DataFrame({"stop_date": [str(a)[:10]], "stop_id": [mean_value]})
        stops_by_day = insert_row(i, stops_by_day, aa)

# finally, remove the column 'new_date' and display the appended dataframe with missing dates
stops_by_day = stops_by_day.drop(columns = ['new_date'])
stops_by_day.head()
````
